```python
import pandas as pd
import numpy as np
from typing import List, Tuple, Dict
from sklearn.cluster import HDBSCAN
import xgboost as xgb
from tqdm import tqdm
import joblib


class DateRangeScaler:
    """
    Standard scaler that fits on a specific date range to avoid lookahead bias.
    Stores mean and std per feature, replacing zero std with 1 to prevent division by zero.
    """
    def __init__(self):
        self.means = None
        self.stds = None
        self.fitted = False

    def fit(self, df: pd.DataFrame, feature_cols: List[str]):
        """
        Compute per-feature mean and standard deviation from the given dataframe slice.
        Zero-variance columns get std=1 to avoid NaN during division.
        """
        self.means = df[feature_cols].mean()
        self.stds = df[feature_cols].std().replace(0, 1)
        self.fitted = True

    def transform(self, df: pd.DataFrame, feature_cols: List[str]) -> pd.DataFrame:
        """
        Apply z-score normalization using stored means and stds.
        Creates a copy to avoid mutating the original dataframe.
        """
        if not self.fitted:
            raise ValueError("Scaler not fitted")

        df_scaled = df.copy()
        df_scaled[feature_cols] = (df[feature_cols] - self.means) / self.stds
        return df_scaled

    def fit_transform(self, df: pd.DataFrame, feature_cols: List[str]) -> pd.DataFrame:
        """
        Convenience method combining fit and transform in one call.
        Fits on the provided data then immediately transforms it.
        """
        self.fit(df, feature_cols)
        return self.transform(df, feature_cols)


class DataManager:
    """
    Central data handler that manages date filtering, feature access, and scaling.
    Maintains integer encoding of stock identifiers and enforces horizon cutoff
    to prevent future information leakage during training.
    """
    def __init__(
        self,
        df: pd.DataFrame,
        feature_cols: List[str],
        delay_cols: List[str] = ["1_w_delay", "1_m_delay", "target_return"],
        date_col: str = "TRADEDATE",
        id_col: str = "SECID",
        use_scaler: bool = False,
        horizon_days: int = 180,
    ):  
        self.df = df.copy()
        # Create bidirectional mapping between ticker strings and integer IDs
        self.secid2id = {s: i for i, s in enumerate(sorted(self.df["SECID"].unique()))}
        self.id2secid = {i: s for s, i in self.secid2id.items()}
        # Preserve original string SECID before replacing with integer encoding
        self.df["SECID_raw"] = self.df["SECID"]
        self.df["SECID"] = self.df["SECID"].map(self.secid2id)
        self.feature_cols = feature_cols
        self.delay_cols = delay_cols
        self.date_col = date_col
        self.id_col = id_col
        self.horizon_days = horizon_days

        self.use_scaler = use_scaler
        self.scaler = DateRangeScaler() if use_scaler else None

        # Ensure proper datetime type and chronological ordering for time-series operations
        self.df[self.date_col] = pd.to_datetime(self.df[self.date_col])
        self.df = self.df.sort_values([self.date_col, self.id_col]).reset_index(drop=True)

        self.available_dates = sorted(self.df[self.date_col].unique())

    def fit_scaler(self, start_date: str, end_date: str):
        """
        Fit scaler on data within specified date range.
        Fitting only on in-sample period prevents lookahead bias
        from leaking future distribution information into normalization.
        """
        if not self.use_scaler:
            return

        mask = (self.df[self.date_col] >= start_date) & (self.df[self.date_col] <= end_date)
        df_slice = self.df.loc[mask]

        self.scaler.fit(df_slice, self.feature_cols)

    def apply_scaler(self):
        """
        Apply fitted scaler to entire dataset.
        Must be called after fit_scaler to have valid normalization parameters.
        """
        if not self.use_scaler:
            return

        self.df = self.scaler.transform(self.df, self.feature_cols)

    def _get_nearest_available_date(self, target_date: pd.Timestamp) -> pd.Timestamp:
        """
        Returns the nearest available date <= target_date.
        Used to handle weekends and holidays when exact calendar dates are missing.
        Returns None if no valid past date exists (e.g., before data starts).
        """
        valid_dates = [d for d in self.available_dates if d <= target_date]
        if not valid_dates:
            return None
        return max(valid_dates)

    def get_snapshot_with_lag(
        self,
        reference_date: str,
        lag_days: int,
    ) -> pd.DataFrame:
        """
        Returns data for reference_date minus lag_days using nearest available date,
        accounting for horizon to avoid lookahead bias.
        The horizon cutoff ensures we don't use data that would need future returns
        beyond what's available in the dataset.
        """
        reference_date = pd.to_datetime(reference_date)
        target_date = reference_date - pd.Timedelta(days=lag_days)

        nearest_date = self._get_nearest_available_date(target_date)
        if nearest_date is None:
            return pd.DataFrame()

        max_date = self.df[self.date_col].max()
        cutoff_date = max_date - pd.Timedelta(days=self.horizon_days)

        # Exclude dates that fall within the horizon buffer zone
        if nearest_date > cutoff_date:
            return pd.DataFrame()

        df_slice = self.df[self.df[self.date_col] == nearest_date]

        cols = [self.id_col, self.date_col] + self.feature_cols + self.delay_cols
        return df_slice[cols].dropna()

    def get_multi_horizon_snapshots(
        self,
        reference_date: str,
        lags: List[int] = [14, 42, 182],
    ) -> Dict[int, pd.DataFrame]:
        """
        Returns data snapshots for multiple lag horizons.
        Useful for generating training examples at different lookback periods
        from a single reference date.
        """
        result = {}

        for lag in lags:
            result[lag] = self.get_snapshot_with_lag(reference_date, lag)

        return result

    def get_training_data(
        self,
        start_date: str,
        end_date: str,
        target_col: str = "target_return",
    ) -> Tuple[np.ndarray, np.ndarray]:
        """
        Returns training data filtered by date range and horizon cutoff.
        The horizon cutoff ensures we never train on dates where the target
        would require future information beyond the available data.
        Drops rows with missing features or targets for clean training.
        """
        max_date = self.df[self.date_col].max()
        cutoff_date = max_date - pd.Timedelta(days=self.horizon_days)

        mask = (
            (self.df[self.date_col] >= start_date) &
            (self.df[self.date_col] <= end_date) &
            (self.df[self.date_col] <= cutoff_date)
        )

        df_slice = self.df.loc[mask]
        df_slice = df_slice.dropna(subset=self.feature_cols + [target_col])

        X = df_slice[self.feature_cols].values
        y = df_slice[target_col].values

        return X, y

    def get_available_dates(self) -> List[pd.Timestamp]:
        """
        Return all unique trading dates in the dataset.
        """
        return self.available_dates

    def get_features_by_date(self, date: str) -> pd.DataFrame:
        """
        Retrieve feature matrix for all stocks on a specific date.
        Used during prediction to get the current snapshot of the market.
        Returns a copy to prevent accidental mutation of the main dataframe.
        """
        date = pd.to_datetime(date)
        return self.df[self.df[self.date_col] == date][
            [self.id_col] + self.feature_cols
        ].copy()

    def decode_secid(self, secid_encoded):
        """
        Convert integer-encoded SECID back to original ticker string.
        Returns the input unchanged if not found in the mapping.
        """
        return self.id2secid.get(secid_encoded, secid_encoded)


class XGBModelWrapper:
    """
    Thin wrapper around XGBoost regressor with fixed hyperparameters.
    Provides fit/predict interface with empty-data guards for edge cases.
    Uses conservative parameters to reduce overfitting risk.
    """
    def __init__(self):
        # low learning_rate + moderate depth + regularization via reg_lambda
        # subsample and colsample_bytree add stochasticity for robustness
        self.model = xgb.XGBRegressor(
            n_estimators=300,
            max_depth=3,
            learning_rate=0.05,
            subsample=0.8,
            colsample_bytree=0.8,
            reg_lambda=1.0,
            random_state=42,
        )

    def fit(self, X, y):
        """
        Train the model if data is non-empty. Silent skip protects
        against edge cases with no training samples for a given period.
        """
        if len(X) == 0:
            return
        self.model.fit(X, y)

    def predict(self, X):
        """
        Generate predictions, returning empty array if no input data.
        Prevents downstream errors when a date has zero valid stocks.
        """
        if len(X) == 0:
            return np.array([])
        return self.model.predict(X)


class EnsembleOrchestrator:
    """
    Manages multi-horizon ensemble of XGBoost models with dynamic weighting.
    Weights are inverse-proportional to historical mean absolute error per horizon.
    Uses delayed feedback: prediction errors are only computed once the true target
    becomes available (after the horizon has fully elapsed).
    In this single-horizon configuration, the ensemble reduces to one model
    but the weighting infrastructure remains for potential multi-horizon extension.
    """
    def __init__(
        self,
        data_manager,
        start_train_date: str,
        start_test_date: str,
        end_date: str,
        train_window_days: int = 365 * 3,
        step_days: int = 14,
        horizons: List[int] = [14, 42, 182],
    ):
        self.dm = data_manager

        self.start_train_date = pd.to_datetime(start_train_date)
        self.start_test_date = pd.to_datetime(start_test_date)
        self.end_date = pd.to_datetime(end_date)

        self.train_window_days = train_window_days
        self.step_days = step_days
        self.horizons = horizons

        # Separate model instance per horizon (here only one: 182 days)
        self.models = {h: XGBModelWrapper() for h in horizons}

        # Registry stores historical predictions for delayed error computation
        self.pred_registry = []
        self.errors = {h: [] for h in horizons}

    def _compute_weights(self):
        """
        Compute ensemble weights based on historical prediction errors.
        Each model's weight is proportional to 1/(mean_error + epsilon),
        normalized to sum to 1. Models with no error history receive equal weight.
        With a single horizon, weight will always be 1.0.
        """
        weights = {}
        for h in self.horizons:
            errs = self.errors[h]
            if len(errs) == 0:
                weights[h] = 1.0
            else:
                # Epsilon prevents division by zero when errors are perfect
                weights[h] = 1.0 / (np.mean(errs) + 1e-6)

        s = sum(weights.values())
        return {k: v / s for k, v in weights.items()}

    def _train_models(self, current_date: pd.Timestamp):
        """
        Train models for all horizons using expanding window.
        The training window starts train_window_days before current_date,
        allowing the model to adapt to recent market conditions.
        Uses a single target_return column for the fixed 180-day horizon.
        """
        train_start = current_date - pd.Timedelta(days=self.train_window_days)

        for h in self.horizons:
            X, y = self.dm.get_training_data(
                start_date=str(train_start.date()),
                end_date=str(current_date.date()),
                target_col="target_return",
            )

            self.models[h].fit(X, y)

    def predict(self, df_day: pd.DataFrame):
        """
        Generate ensemble prediction using weighted average of individual models.
        For the single-horizon case, this is simply the 182-day model output.
        Returns a flat numpy array of predictions aligned with input rows.
        """
        if df_day.empty:
            return np.array([])

        X = df_day[self.dm.feature_cols].values

        # Collect predictions from each horizon model
        preds_by_model = {}
        for h in self.horizons:
            preds_by_model[h] = self.models[h].predict(X)

        weights = self._compute_weights()

        # Weighted linear combination (single model gets full weight of 1.0)
        ensemble_pred = np.zeros(len(X))
        for h in self.horizons:
            ensemble_pred += weights[h] * preds_by_model[h]

        return ensemble_pred

    def _update_errors(self, current_date: pd.Timestamp):
        """
        Update prediction errors using delayed feedback mechanism.
        For each past prediction, checks whether enough calendar time has elapsed
        (>= horizon_days) for the true target to be known. If so, computes MAE
        and appends to the error history for all horizons.
        In this version, errors are shared across horizons since there's only
        one target column.
        """
        for record in self.pred_registry:
            pred_date = record["date"]

            # Only evaluate if the horizon period has fully elapsed
            if (current_date - pred_date).days < self.dm.horizon_days:
                continue

            df_true_full = self.dm.df[self.dm.df[self.dm.date_col] == current_date]

            if df_true_full.empty:
                continue

            y_true = df_true_full[["SECID", "target_return"]].values
            ids_pred = record["ids"]
            y_pred = record["preds"]

            if len(y_true) != len(y_pred):
                continue

            # Align predictions with realized returns using SECID mapping
            true_map = {row[0]: row[1] for row in y_true}

            aligned_true = np.array([true_map.get(secid, np.nan) for secid in ids_pred])

            mask = ~np.isnan(aligned_true)

            # Mean absolute error on matched stocks only
            err = np.mean(np.abs(aligned_true[mask] - y_pred[mask]))

            for h in self.horizons:
                self.errors[h].append(err)

    def run(self):
        """
        Execute main training and prediction loop.
        Walks forward from start_test_date to end_date with step_days increments,
        training models at each step and accumulating predictions for error tracking.
        """
        current_date = self.start_test_date

        while current_date <= self.end_date:
            print(f"\n[{current_date.date()}]")

            self._train_models(current_date)
            self._predict(current_date)
            self._update_errors(current_date)

            weights = self._compute_weights()
            print("weights:", weights)
            print("errors:", {k: np.mean(v) if v else None for k, v in self.errors.items()})

            current_date += pd.Timedelta(days=self.step_days)

    def get_last_errors(self) -> Dict[int, float]:
        """
        Returns the last recorded error for each horizon.
        Useful for quick inspection of current model performance.
        Returns None for horizons with no error history yet.
        """
        last_errors = {}
        
        for h in self.horizons:
            if len(self.errors[h]) > 0:
                last_errors[h] = self.errors[h][-1]
            else:
                last_errors[h] = None
        
        return last_errors


def portfolio_pipeline(
    df_day: pd.DataFrame,
    model,
    feature_cols,
    data_manager,
    secid_col="SECID",
    top_k_small=2,
    top_k_large=3,
):
    """
    Construct portfolio from daily predictions using clustering and top selection.
    First filters to stocks with positive ensemble predictions only.
    Then applies HDBSCAN clustering on feature space to group similar stocks.
    From each cluster, selects the top-k stocks by prediction strength,
    with larger clusters (size > 5) getting more slots for better diversification.
    Final weights are proportional to prediction values within the selected set.
    """
    df = df_day.copy()

    print("rows before filter:", len(df))
    print("rows after filter:", len(df[df["pred"] > 0]))
    # Keep only stocks with positive expected return signal
    df = df[df["pred"] > 0]

    if df.empty:
        return pd.DataFrame()

    # HDBSCAN with Euclidean distance groups stocks with similar feature profiles
    # min_cluster_size=3 prevents micro-clusters, eom method selects stable clusters
    clusterer = HDBSCAN(
        min_cluster_size=3,
        min_samples=3,
        metric="euclidean",
        cluster_selection_method="eom"
    )
    # Exclude the identifier column from clustering features
    numeric_features = [c for c in feature_cols if c != secid_col]
    df["cluster"] = clusterer.fit_predict(df[numeric_features].values)
    if df.empty:
        return pd.DataFrame()

    selected = []
    print("clusters:", np.unique(df["cluster"], return_counts=True))
    for c in df["cluster"].unique():
        cluster_df = df[df["cluster"] == c]
        cluster_df = cluster_df.sort_values("pred", ascending=False)

        # Larger clusters get more representatives for better diversification
        k = top_k_large if len(cluster_df) > 5 else top_k_small
        selected.append(cluster_df.head(k))

    selected = pd.concat(selected)

    # Weights proportional to prediction strength, sum to 1
    selected["weight"] = selected["pred"] / selected["pred"].sum()

    selected = selected.copy()
    # Convert back from integer encoding to original ticker strings for output
    selected["SECID"] = selected["SECID"].apply(data_manager.id2secid.get)

    return selected[["SECID", "pred", "weight", "cluster"]]


def evaluate_portfolio(portfolio, date, data_manager):
    """
    Calculate realized portfolio return at the horizon date.
    Looks up the nearest available trading date at date + horizon_days,
    then computes the weighted sum of actual forward returns for the selected stocks.
    Returns both the portfolio-level return and the per-stock detail dataframe.
    """
    horizon = data_manager.horizon_days
    future_date = pd.to_datetime(date) + pd.Timedelta(days=horizon)

    # Use nearest available date because exact horizon date may be a holiday/weekend
    future_date = data_manager._get_nearest_available_date(future_date)

    if future_date is None:
        print("No future data")
        return None

    df_future = data_manager.df[
        data_manager.df[data_manager.date_col] == future_date
    ][["SECID", "target_return"]].copy()

    # Map integer SECID back to string for compatibility with portfolio ticker strings
    df_future["SECID"] = df_future["SECID"].map(data_manager.id2secid)

    merged = portfolio.merge(df_future, on="SECID", how="left")

    # Drop stocks that disappeared before the horizon date (no return data available)
    merged = merged.dropna(subset=["target_return"])

    if merged.empty:
        print("No matched future returns")
        return None

    # Weighted sum gives the portfolio-level realized return
    portfolio_return = (merged["weight"] * merged["target_return"]).sum()

    return portfolio_return, merged


def run_backtest(
    df,
    feature_cols,
    data_manager,
    start_train_date,
    start_test_date,
    end_date,
    portfolio_pipeline_fn,
    save_path="ensemble_180.pkl",
    step_days=14,
    portfolio_dates=None,
):
    """
    Execute full backtest: training loop followed by portfolio construction and evaluation.
    The training phase walks forward in time with step_days intervals, training
    expanding-window models at each step and recording predictions for delayed
    error computation using the 180-day target return.
    After training completes, the trained orchestrator is saved to disk.
    Then separately evaluates specific portfolio dates by running the pipeline and
    computing realized returns against the horizon.
    """
    # Initialize orchestrator with single 182-day horizon for this configuration
    orchestrator = EnsembleOrchestrator(
        data_manager=data_manager,
        start_train_date=start_train_date,
        start_test_date=start_test_date,
        end_date=end_date,
        horizons=[182],
        step_days=step_days,
    )

    dates = pd.date_range(start_test_date, end_date, freq=f"{step_days}D")

    # Forward walk: train models and accumulate prediction registry for error tracking
    for current_date in tqdm(dates, desc="Training"):
        print(f"\nTRAIN DATE: {current_date.date()}")

        orchestrator._train_models(current_date)

        df_day = data_manager.get_features_by_date(current_date)
        if not df_day.empty:
            preds = orchestrator.predict(df_day)

            # Store predictions and stock IDs for later error computation
            # when true 180-day forward returns become available
            orchestrator.pred_registry.append({
                "date": current_date,
                "ids": df_day["SECID"].values,
                "preds": preds,
            })

        # Check if any past predictions have matured and can be evaluated
        orchestrator._update_errors(current_date)
    
    # Display final error metrics before saving
    last_errors = {h: errs[-1] if errs else None for h, errs in orchestrator.errors.items()}
    print(f"Last Errors:")
    for h, err in last_errors.items():
        print(f"  Horizon {h}: {err}")
    
    # Persist the trained orchestrator for later use or inspection
    joblib.dump(orchestrator, save_path)
    print(f"\nMODEL SAVED -> {save_path}")

    portfolios = []

    # Portfolio evaluation phase: run on specific test dates using the trained orchestrator
    if portfolio_dates is not None:
        print("\n=== PORTFOLIO TEST ===")

        for d in tqdm(portfolio_dates, desc="Portfolio"):
            d = pd.to_datetime(d)
            print(f"\nPORTFOLIO DATE: {d.date()}")

            df_day = data_manager.get_features_by_date(d)

            if df_day.empty:
                print("No data for this date.")
                continue

            preds = orchestrator.predict(df_day)

            df_day = df_day.copy()
            df_day["pred"] = preds

            portfolio = portfolio_pipeline_fn(
                df_day=df_day,
                model=None,
                feature_cols=feature_cols,
                data_manager=data_manager,
            )
            print(f"Selected {len(portfolio)} stocks for portfolio.")

            if not portfolio.empty:
                portfolios.append({
                    "date": d,
                    "portfolio": portfolio
                })

                print(portfolio.head())
                result = evaluate_portfolio(portfolio, d, data_manager)

                if result is not None:
                    port_ret, details = result

                    print(portfolio.head())
                    print("\nPortfolio return:", round(port_ret, 4))
                    print("Details:")
                    print(details.head())

                    print(portfolios)


if __name__ == "__main__":
    # Load raw dataset with automatic date parsing for the trading date column
    df = pd.read_csv('output_with_cbr_rate.csv', parse_dates=['TRADEDATE'])

    # Feature set combines momentum, volatility, risk, liquidity, dividend,
    # relative strength, beta, size, and macro rate signals
    feature_cols = [
        "momentum_6m", "momentum_12m", "momentum_24m", "momentum_36m",
        "volatility_6m", "volatility_12m", "volatility_24m", "volatility_36m",
        "max_drawdown_6m", "max_drawdown_12m", "max_drawdown_24m", "max_drawdown_36m",
        "avg_value_6m", "avg_value_12m", "avg_value_36m",
        "dividend_yield_12m", "dividend_yield_3y_avg", "dividend_consistency_3y",
        "relative_strength_12m", "excess_return_6m", "beta_12m",
        "log_issuesize", "cbr_key_rate"
    ]

    # Initialize data manager with 180-day horizon and single target_return column
    # Scaler is fitted on 2015-2018 to establish normalization baseline
    dm = DataManager(
        df=df,
        feature_cols=feature_cols,
        delay_cols=["target_return"],
        date_col="TRADEDATE",
        id_col="SECID",
        use_scaler=True,
        horizon_days=180,
    )

    # Fit scaler on in-sample period to avoid lookahead, then transform full dataset
    dm.fit_scaler(start_date="2015-01-01", end_date="2018-01-01")
    dm.apply_scaler()

    # Run backtest with training from 2015 to 2024, evaluate on specific recent dates
    # Uses single 182-day horizon model for simplicity
    history = run_backtest(
        df=df,
        feature_cols=feature_cols,
        data_manager=dm,
        start_train_date="2015-01-01",
        start_test_date="2018-01-01",
        end_date="2024-01-01",
        portfolio_pipeline_fn=portfolio_pipeline,
        portfolio_dates=[
            "2024-08-05",
            "2024-09-03",
            "2025-04-03",
            "2025-05-05",
            "2025-06-06"
        ]
    )
```

```python
import pandas as pd
import numpy as np
from typing import List, Tuple, Dict
from sklearn.cluster import HDBSCAN
import xgboost as xgb
from tqdm import tqdm
import joblib


class DateRangeScaler:
    """
    Standard scaler that fits on a specific date range to avoid lookahead bias.
    Stores mean and std per feature, replacing zero std with 1 to prevent division by zero.
    """
    def __init__(self):
        self.means = None
        self.stds = None
        self.fitted = False

    def fit(self, df: pd.DataFrame, feature_cols: List[str]):
        """
        Compute per-feature mean and standard deviation from the given dataframe slice.
        Zero-variance columns get std=1 to avoid NaN during division.
        """
        self.means = df[feature_cols].mean()
        self.stds = df[feature_cols].std().replace(0, 1)
        self.fitted = True

    def transform(self, df: pd.DataFrame, feature_cols: List[str]) -> pd.DataFrame:
        """
        Apply z-score normalization using stored means and stds.
        Creates a copy to avoid mutating the original dataframe.
        """
        if not self.fitted:
            raise ValueError("Scaler not fitted")

        df_scaled = df.copy()
        df_scaled[feature_cols] = (df[feature_cols] - self.means) / self.stds
        return df_scaled

    def fit_transform(self, df: pd.DataFrame, feature_cols: List[str]) -> pd.DataFrame:
        """
        Convenience method combining fit and transform in one call.
        Fits on the provided data then immediately transforms it.
        """
        self.fit(df, feature_cols)
        return self.transform(df, feature_cols)


class DataManager:
    """
    Central data handler that manages date filtering, feature access, and scaling.
    Maintains integer encoding of stock identifiers and enforces horizon cutoff
    to prevent future information leakage during training.
    """
    def __init__(
        self,
        df: pd.DataFrame,
        feature_cols: List[str],
        delay_cols: List[str] = ["1_w_delay", "1_m_delay", "target_return"],
        date_col: str = "TRADEDATE",
        id_col: str = "SECID",
        use_scaler: bool = False,
        horizon_days: int = 180,
    ):  
        self.df = df.copy()
        # Create bidirectional mapping between ticker strings and integer IDs
        # for efficient storage and fast lookup operations
        self.secid2id = {s: i for i, s in enumerate(sorted(self.df["SECID"].unique()))}
        self.id2secid = {i: s for s, i in self.secid2id.items()}
        # Preserve original string SECID before replacing with integer encoding
        self.df["SECID_raw"] = self.df["SECID"]
        self.df["SECID"] = self.df["SECID"].map(self.secid2id)
        self.feature_cols = feature_cols
        self.delay_cols = delay_cols
        self.date_col = date_col
        self.id_col = id_col
        self.horizon_days = horizon_days

        self.use_scaler = use_scaler
        self.scaler = DateRangeScaler() if use_scaler else None

        # Ensure proper datetime type and chronological ordering for time-series operations
        self.df[self.date_col] = pd.to_datetime(self.df[self.date_col])
        self.df = self.df.sort_values([self.date_col, self.id_col]).reset_index(drop=True)

        self.available_dates = sorted(self.df[self.date_col].unique())

    def fit_scaler(self, start_date: str, end_date: str):
        """
        Fit scaler on data within specified date range.
        Fitting only on in-sample period prevents lookahead bias
        from leaking future distribution information into normalization.
        """
        if not self.use_scaler:
            return

        mask = (self.df[self.date_col] >= start_date) & (self.df[self.date_col] <= end_date)
        df_slice = self.df.loc[mask]

        self.scaler.fit(df_slice, self.feature_cols)

    def apply_scaler(self):
        """
        Apply fitted scaler to entire dataset.
        Must be called after fit_scaler to have valid normalization parameters.
        """
        if not self.use_scaler:
            return

        self.df = self.scaler.transform(self.df, self.feature_cols)

    def _get_nearest_available_date(self, target_date: pd.Timestamp) -> pd.Timestamp:
        """
        Returns the nearest available date <= target_date.
        Used to handle weekends and holidays when exact calendar dates are missing.
        Returns None if no valid past date exists (e.g., before data starts).
        """
        valid_dates = [d for d in self.available_dates if d <= target_date]
        if not valid_dates:
            return None
        return max(valid_dates)

    def get_snapshot_with_lag(
        self,
        reference_date: str,
        lag_days: int,
    ) -> pd.DataFrame:
        """
        Returns data for reference_date minus lag_days using nearest available date,
        accounting for horizon to avoid lookahead bias.
        The horizon cutoff ensures we don't use data that would need future returns
        beyond what's available in the dataset.
        """
        reference_date = pd.to_datetime(reference_date)
        target_date = reference_date - pd.Timedelta(days=lag_days)

        nearest_date = self._get_nearest_available_date(target_date)
        if nearest_date is None:
            return pd.DataFrame()

        max_date = self.df[self.date_col].max()
        cutoff_date = max_date - pd.Timedelta(days=self.horizon_days)

        # Exclude dates that fall within the horizon buffer zone
        # to prevent using data whose target hasn't materialized yet
        if nearest_date > cutoff_date:
            return pd.DataFrame()

        df_slice = self.df[self.df[self.date_col] == nearest_date]

        cols = [self.id_col, self.date_col] + self.feature_cols + self.delay_cols
        return df_slice[cols].dropna()

    def get_multi_horizon_snapshots(
        self,
        reference_date: str,
        lags: List[int] = [14, 42, 182],
    ) -> Dict[int, pd.DataFrame]:
        """
        Returns data snapshots for multiple lag horizons.
        Useful for generating training examples at different lookback periods
        from a single reference date.
        """
        result = {}

        for lag in lags:
            result[lag] = self.get_snapshot_with_lag(reference_date, lag)

        return result

    def get_training_data(
        self,
        start_date: str,
        end_date: str,
        target_col: str = "target_return",
    ) -> Tuple[np.ndarray, np.ndarray]:
        """
        Returns training data filtered by date range and horizon cutoff.
        The horizon cutoff ensures we never train on dates where the target
        would require future information beyond the available data.
        Drops rows with missing features or targets for clean training.
        """
        max_date = self.df[self.date_col].max()
        cutoff_date = max_date - pd.Timedelta(days=self.horizon_days)

        mask = (
            (self.df[self.date_col] >= start_date) &
            (self.df[self.date_col] <= end_date) &
            (self.df[self.date_col] <= cutoff_date)
        )

        df_slice = self.df.loc[mask]
        # Remove rows where features or target are missing to ensure clean training data
        df_slice = df_slice.dropna(subset=self.feature_cols + [target_col])

        X = df_slice[self.feature_cols].values
        y = df_slice[target_col].values

        return X, y

    def get_available_dates(self) -> List[pd.Timestamp]:
        """
        Return all unique trading dates in the dataset sorted chronologically.
        """
        return self.available_dates

    def get_features_by_date(self, date: str) -> pd.DataFrame:
        """
        Retrieve feature matrix for all stocks on a specific date.
        Used during prediction to get the current snapshot of the market.
        Returns a copy to prevent accidental mutation of the main dataframe.
        """
        date = pd.to_datetime(date)
        return self.df[self.df[self.date_col] == date][
            [self.id_col] + self.feature_cols
        ].copy()

    def decode_secid(self, secid_encoded):
        """
        Convert integer-encoded SECID back to original ticker string.
        Returns the input unchanged if not found in the mapping.
        """
        return self.id2secid.get(secid_encoded, secid_encoded)


class XGBModelWrapper:
    """
    Thin wrapper around XGBoost regressor with fixed hyperparameters.
    Provides fit/predict interface with empty-data guards for edge cases.
    Uses conservative parameters to reduce overfitting risk on financial data.
    """
    def __init__(self):
        # Conservative hyperparameters: low learning rate, shallow trees,
        # row/column subsampling for regularization, L2 penalty via reg_lambda
        self.model = xgb.XGBRegressor(
            n_estimators=300,
            max_depth=3,
            learning_rate=0.05,
            subsample=0.8,
            colsample_bytree=0.8,
            reg_lambda=1.0,
            random_state=42,
        )

    def fit(self, X, y):
        """
        Train the model if data is non-empty. Silent skip protects
        against edge cases with no training samples for a given period.
        """
        if len(X) == 0:
            return
        self.model.fit(X, y)

    def predict(self, X):
        """
        Generate predictions, returning empty array if no input data.
        Prevents downstream errors when a date has zero valid stocks.
        """
        if len(X) == 0:
            return np.array([])
        return self.model.predict(X)


class EnsembleOrchestrator:
    """
    Manages multi-horizon ensemble of XGBoost models with dynamic weighting.
    Weights are inverse-proportional to historical mean absolute error per horizon.
    Uses delayed feedback: prediction errors are only computed once the true target
    becomes available (after the horizon has fully elapsed).
    In single-horizon configuration, the ensemble reduces to one model
    but the weighting infrastructure remains for potential multi-horizon extension.
    """
    def __init__(
        self,
        data_manager,
        start_train_date: str,
        start_test_date: str,
        end_date: str,
        train_window_days: int = 365 * 3,
        step_days: int = 14,
        horizons: List[int] = [14, 42, 182],
    ):
        self.dm = data_manager

        self.start_train_date = pd.to_datetime(start_train_date)
        self.start_test_date = pd.to_datetime(start_test_date)
        self.end_date = pd.to_datetime(end_date)

        self.train_window_days = train_window_days
        self.step_days = step_days
        self.horizons = horizons

        # Separate model instance per horizon captures different return frequencies
        self.models = {h: XGBModelWrapper() for h in horizons}

        # Registry stores historical predictions for delayed error computation
        self.pred_registry = []
        self.errors = {h: [] for h in horizons}

    def _compute_weights(self):
        """
        Compute ensemble weights based on historical prediction errors.
        Each model's weight is proportional to 1/(mean_error + epsilon),
        normalized to sum to 1. Models with no error history receive equal weight.
        With a single horizon, weight will always be 1.0.
        """
        weights = {}
        for h in self.horizons:
            errs = self.errors[h]
            if len(errs) == 0:
                weights[h] = 1.0
            else:
                # Epsilon prevents division by zero when errors are perfect
                weights[h] = 1.0 / (np.mean(errs) + 1e-6)

        s = sum(weights.values())
        return {k: v / s for k, v in weights.items()}

    def _train_models(self, current_date: pd.Timestamp):
        """
        Train models for all horizons using expanding window.
        The training window starts train_window_days before current_date,
        allowing the model to adapt to recent market conditions.
        Uses a single target_return column regardless of horizon count.
        """
        train_start = current_date - pd.Timedelta(days=self.train_window_days)

        for h in self.horizons:
            X, y = self.dm.get_training_data(
                start_date=str(train_start.date()),
                end_date=str(current_date.date()),
                target_col="target_return",
            )

            self.models[h].fit(X, y)

    def predict(self, df_day: pd.DataFrame):
        """
        Generate ensemble prediction using weighted average of individual models.
        For the single-horizon case, this is simply the model output.
        Returns a flat numpy array of predictions aligned with input rows.
        """
        if df_day.empty:
            return np.array([])

        X = df_day[self.dm.feature_cols].values

        # Collect predictions from each horizon model
        preds_by_model = {}
        for h in self.horizons:
            preds_by_model[h] = self.models[h].predict(X)

        weights = self._compute_weights()

        # Weighted linear combination of all horizon predictions
        ensemble_pred = np.zeros(len(X))
        for h in self.horizons:
            ensemble_pred += weights[h] * preds_by_model[h]

        return ensemble_pred

    def _update_errors(self, current_date: pd.Timestamp):
        """
        Update prediction errors using delayed feedback mechanism.
        For each past prediction, checks whether enough calendar time has elapsed
        (>= horizon_days) for the true target to be known. If so, computes MAE
        and appends to the error history for all horizons.
        Errors are shared across horizons since there's only one target column.
        """
        for record in self.pred_registry:
            pred_date = record["date"]

            # Only evaluate if the horizon period has fully elapsed
            if (current_date - pred_date).days < self.dm.horizon_days:
                continue

            df_true_full = self.dm.df[self.dm.df[self.dm.date_col] == current_date]

            if df_true_full.empty:
                continue

            y_true = df_true_full[["SECID", "target_return"]].values
            ids_pred = record["ids"]
            y_pred = record["preds"]

            if len(y_true) != len(y_pred):
                continue

            # Align predictions with realized returns using SECID mapping
            true_map = {row[0]: row[1] for row in y_true}

            aligned_true = np.array([true_map.get(secid, np.nan) for secid in ids_pred])

            mask = ~np.isnan(aligned_true)

            # Mean absolute error computed only on stocks that survived to horizon
            err = np.mean(np.abs(aligned_true[mask] - y_pred[mask]))

            for h in self.horizons:
                self.errors[h].append(err)

    def run(self):
        """
        Execute main training and prediction loop.
        Walks forward from start_test_date to end_date with step_days increments,
        training models at each step and accumulating predictions for error tracking.
        Prints current weights and error metrics at each iteration.
        """
        current_date = self.start_test_date

        while current_date <= self.end_date:
            print(f"\n[{current_date.date()}]")

            self._train_models(current_date)
            self._predict(current_date)
            self._update_errors(current_date)

            weights = self._compute_weights()
            print("weights:", weights)
            print("errors:", {k: np.mean(v) if v else None for k, v in self.errors.items()})

            current_date += pd.Timedelta(days=self.step_days)

    def get_last_errors(self) -> Dict[int, float]:
        """
        Returns the last recorded error for each horizon.
        Useful for quick inspection of current model performance.
        Returns None for horizons with no error history yet.
        """
        last_errors = {}
        
        for h in self.horizons:
            if len(self.errors[h]) > 0:
                last_errors[h] = self.errors[h][-1]
            else:
                last_errors[h] = None
        
```

```python
import pickle
import joblib
import matplotlib.pyplot as plt
import numpy as np

def load_model_and_visualize(model_path="ensemble_180.pkl", feature_cols=None):
    """
    Load a saved orchestrator from disk and visualize its internals.
    Displays feature importance per horizon model and ensemble weighting scheme.
    Returns the loaded orchestrator for further inspection if needed.
    """
    print(f"Loading model from {model_path}...")
    orchestrator = joblib.load(model_path)
    print("Model loaded successfully!")
    
    # Default feature list matching the training setup; used to label importance plots
    if feature_cols is None:
        feature_cols = [
            "momentum_6m", "momentum_12m", "momentum_24m", "momentum_36m",
            "volatility_6m", "volatility_12m", "volatility_24m", "volatility_36m",
            "max_drawdown_6m", "max_drawdown_12m", "max_drawdown_24m", "max_drawdown_36m",
            "avg_value_6m", "avg_value_12m", "avg_value_36m",
            "dividend_yield_12m", "dividend_yield_3y_avg", "dividend_consistency_3y",
            "relative_strength_12m", "excess_return_6m", "beta_12m",
            "log_issuesize", "cbr_key_rate"
        ]
    
    visualize_model(orchestrator, feature_cols)
    
    return orchestrator

def visualize_model(orchestrator, feature_cols):
    """
    Generate three diagnostic outputs for model inspection:
    1. Per-horizon feature importance (top 15 features) from each XGBoost model
    2. Ensemble weights bar chart showing how much each horizon contributes
    3. Console summary of errors and weights
    
    The feature importance reveals which signals each horizon model relies on,
    helping diagnose whether short-term models focus on momentum while
    long-term models emphasize value/fundamental features.
    """
    n_models = len(orchestrator.horizons)
    # Create side-by-side subplots for each horizon's feature importance
    fig, axes = plt.subplots(1, n_models, figsize=(5*n_models, 6))
    
    # Handle single-horizon case gracefully: wrap in list for consistent indexing
    if n_models == 1:
        axes = [axes]
    
    for idx, horizon in enumerate(orchestrator.horizons):
        # Access the underlying XGBoost model through the wrapper to get importance scores
        model = orchestrator.models[horizon].model
        importance = model.feature_importances_
        
        # Select top 15 most important features for readability in the plot
        sorted_idx = np.argsort(importance)[::-1][:15]
        sorted_importance = importance[sorted_idx]
        sorted_features = [feature_cols[i] for i in sorted_idx]
        
        # Horizontal bar chart with reversed order so most important is at top
        axes[idx].barh(range(len(sorted_idx)), sorted_importance[::-1])
        axes[idx].set_yticks(range(len(sorted_idx)))
        axes[idx].set_yticklabels(sorted_features[::-1], fontsize=8)
        axes[idx].set_xlabel('Importance')
        axes[idx].set_title(f'Horizon {horizon} days')
    
    plt.suptitle('Feature Importance by Horizon', fontsize=14)
    plt.tight_layout()
    plt.show()
    
    # Compute current ensemble weights based on accumulated error history
    weights = orchestrator._compute_weights()
    
    # Bar chart showing relative contribution of each horizon to final prediction
    plt.figure(figsize=(8, 6))
    horizons_str = [f'{h} days' for h in orchestrator.horizons]
    colors = plt.cm.viridis(np.linspace(0, 1, len(horizons_str)))
    
    bars = plt.bar(horizons_str, weights.values(), color=colors)
    plt.ylabel('Weight')
    plt.title('Ensemble Model Weights')
    
    # Annotate each bar with its exact weight value for precise comparison
    for bar, weight in zip(bars, weights.values()):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
                f'{weight:.3f}', ha='center', va='bottom')
    
    plt.ylim(0, max(weights.values()) + 0.1)
    plt.tight_layout()
    plt.show()
    
    # Console summary: key metrics for model evaluation
    print("\n" + "="*50)
    print("MODEL INFORMATION")
    print("="*50)
    print(f"Horizons: {orchestrator.horizons}")
    print(f"\nLast errors for each horizon:")
    for h in orchestrator.horizons:
        if orchestrator.errors[h]:
            # Display the most recent error value to show current model performance
            print(f"  Horizon {h}: {orchestrator.errors[h][-1]:.6f}")
    
    print(f"\nEnsemble weights:")
    for h, w in weights.items():
        print(f"  Horizon {h}: {w:.4f}")

if __name__ == "__main__":
    # Load the saved orchestrator and generate all diagnostic visualizations
    orchestrator = load_model_and_visualize("ensemble_180.pkl")
```
